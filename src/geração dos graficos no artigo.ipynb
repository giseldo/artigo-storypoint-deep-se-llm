```python
import matplotlib.pyplot as plt

# Dados
configuracoes = ["LLM Few-Shot", "LLM Zero-Shot"]
valores = [10, 6]

cores = ["#ff7f0e", "orange"]

# Criando o gráfico de barras
plt.bar(configuracoes, valores, color=cores)

# Rótulos
plt.xlabel("Configuração")
plt.ylabel("Projetos com menor MAE")

plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.grid(axis="x", linestyle="--", alpha=0.7)

# Mostrando o gráfico
plt.show()
```

```python
import matplotlib.pyplot as plt

# Dados
modelos = ["LLM fine-tuning", "LLM Few Shot"]
quantidade = [14, 2]
cores = ["#1f77b4", "#ff7f0e"]

# Criar gráfico
plt.figure(figsize=(6,4))
plt.bar(modelos, quantidade, color=cores)
plt.xlabel("Modelo")
plt.ylabel("Quantidade de Projetos")
plt.title("Comparação de Modelos")
plt.grid(axis="y", linestyle="--", alpha=0.7)

# Mostrar gráfico
plt.show()
```
